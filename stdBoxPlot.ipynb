# Read CSV

```python
import pandas as pd
df_neg = pd.read_csv('neg.csv')
df_pos = pd.read_csv('pos.csv')

col = ['c'+str(i) for i in range(4201)]
df_neg.columns = col
df_pos.columns = col
```

```python
df_neg['mean'] = df_neg.iloc[: , 1:].mean(axis=1)
df_neg["std"] = df_neg.iloc[: , 1:].std(axis=1)

df_pos['mean'] = df_pos.iloc[: , 1:].mean(axis=1)
df_pos["std"] = df_pos.iloc[: , 1:].std(axis=1)
```

```python
metadata = []
labels = df_neg.iloc[:,0].astype(str)
for label in labels:
    name = course_list[label]
    metadata.append(name)
```

# BarPlot

```python
%matplotlib inline

import matplotlib.style as style 
import matplotlib.pyplot as plt
import seaborn as sns
style.use('ggplot')
sns.set_context('paper')
plt.figure(figsize=(10,15))
#plt.rcParams['axes.facecolor'] ='w'
#plt.rcParams['lines.linewidth'] = 3
#plt.rcParams['xtick.color']='r'

size = df_neg.shape[0]
x= range(0, size)
plt.barh(labels, df_pos["mean"], color = 'C5', xerr = df_pos["std"], label = "Attention", alpha=0.85)
plt.barh(labels, df_neg["mean"], color = 'C4', xerr = df_neg["std"], left = df_pos["mean"], label="Non Attention", alpha=0.85)

ax = plt.gca()
#plt.draw()

ax.set_yticklabels([ str(y) for y in metadata])
plt.margins(x =0.04, y=0.005)
#plt.savefig("barplot.jpg",bbox_inches='tight',dpi=1200)
```

# BoxPlot

```python
numerator = df_pos.iloc[: , 1:-2] * 100 # remove filename[0], std[-1] and mean[-2]
denominator = df_pos.iloc[: , 1:-2] + df_neg.iloc[: , 1:-2]
ratio = numerator / denominator
```

```python
%matplotlib inline
import numpy as np
import matplotlib.style as style
import matplotlib.pyplot as plt
import seaborn as sns

#style.use('ggplot')
sns.set_context('paper')
#plt.rcParams['axes.facecolor'] ='w'
plt.figure(figsize=(6, 15))

ax = sns.boxplot(data=ratio.T, orient="h", showfliers = False, palette="Set3")
#ax.axvspan(20, 80, facecolor='#39FF14', alpha=0.05)

#plt.ylim(-1,67)
ax.set_yticklabels([ str(y) for y in metadata])
ax.axvspan(20, 80, facecolor='#39FF14', alpha=0.05)
sns.despine()
#plt.tight_layout()
plt.ylim(-1,55)
#plt.savefig("boxplot.jpg",bbox_inches='tight',dpi=1200)
```
